```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Unemployment in India.csv")
lockdown_start_date = pd.to_datetime("2020-03-25")
lockdown_end_date = pd.to_datetime("2020-05-31")
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.dropna()
df = df.reset_index()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df.info()
```

```python
df["month"] = df["Date"].dt.month
```

```python
df["year"] = df["Date"].dt.year
```

```python
df.head()
```

```python

plt.figure(figsize=(14, 6))
sns.barplot(data = df[(df["Date"]>lockdown_start_date) & (df["Date"] < lockdown_end_date)].groupby("Region")["Estimated Unemployment Rate (%)"].mean().reset_index().sort_values(by="Estimated Unemployment Rate (%)", ascending=False), y = "Region", x = "Estimated Unemployment Rate (%)", errorbar=None)
plt.xticks(rotation = 90)
plt.title("Average Uemployment Percentage Of States During The Lockdown")
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.lineplot(data = df[df["Date"] < "2020-03-25"], x = "Date",y = "Estimated Unemployment Rate (%)", hue = "Area")
plt.title("Before Lockdown Unemployment In Rural And Urban Areas")
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.lineplot(data = df[df["Date"] > "2020-03-25"], x = "Date",y = "Estimated Unemployment Rate (%)", hue = "Area")
plt.title("After Lockdown Unemployment In Rural And Urban Areas")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.axvline(x = lockdown_start_date ,label = "lockdown_start_date")
plt.text(lockdown_start_date,max(df["Estimated Unemployment Rate (%)"])*0.2, s = "Lockdown Start Date", rotation = 90)
plt.axvline(x = lockdown_end_date ,label = "lockdown_end_date")
plt.text(lockdown_end_date,max(df["Estimated Unemployment Rate (%)"])*0.2, s = "Lockdown end Date", rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.lineplot(data = df[(df["Date"] > lockdown_start_date) & (df["Region"].isin({'Puducherry', 'Tamil Nadu', 'Jharkhand', 'Bihar', 'Karnataka',
       'Haryana', 'Kerala', 'Telangana', 'Madhya Pradesh', "Himachal Pradesh",
       'Andhra Pradesh'}))], x = "Date",y = "Estimated Unemployment Rate (%)", hue = "Region")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.axvline(x = lockdown_start_date ,label = "lockdown_start_date")
plt.text(lockdown_start_date,max(df["Estimated Unemployment Rate (%)"])*0.2, s = "Lockdown Start Date", rotation = 90)
plt.axvline(x = lockdown_end_date ,label = "lockdown_end_date")
plt.text(lockdown_end_date,max(df["Estimated Unemployment Rate (%)"])*0.2, s = "Lockdown end Date", rotation = 90)
plt.title("After Lockdown Unemployment In Top 10 Most Affected State")
plt.show()
```

```python

plt.title('Labour Participation vs. Unemployment Rate by State')
sns.scatterplot(data=df, x = "Estimated Labour Participation Rate (%)", y = "Estimated Unemployment Rate (%)", hue = "Region" )
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
sns.regplot(df,  x = "Estimated Labour Participation Rate (%)", y = "Estimated Unemployment Rate (%)")
plt.show()
```

```python
before = df[df["Date"]<lockdown_start_date]
b = before.groupby("Region")["Estimated Unemployment Rate (%)"].mean().reset_index()
b["period"] = "before"

after = df[(df["Date"]>=lockdown_start_date) & (df["Date"] < lockdown_end_date)]
a = after.groupby("Region")["Estimated Unemployment Rate (%)"].mean().reset_index()
a["period"] = "after"


temp = pd.concat([a, b])
temp["diff"] = a["Estimated Unemployment Rate (%)"] - b["Estimated Unemployment Rate (%)"]
temp = temp.sort_values(by = "diff", ascending=False)
temp.head()
```

```python
pd.options.mode.copy_on_write = True 
```

```python
plt.figure(figsize=(14,4))
sns.barplot(data = temp, x = "Region", y = "Estimated Unemployment Rate (%)", hue = "period")
plt.title("Effect Of Lockdown From Low High To Low")
plt.xticks(rotation = 90)
plt.show()
```
